```python
import pandas as pd
```

```python
users = pd.read_csv('users.csv', sep=';').drop_duplicates()
participants = pd.read_csv('participants.csv', sep=';')
lessons = pd.read_csv('lessons.csv', sep=';').drop(['sche', 'duled_time'], axis=1)
quality = pd.read_csv('quality.csv', sep=';')
```

```python
#1. Найти все уроки по физике.
phys = lessons[lessons['subject'] == 'phys']
phys
```

```python
phys_participants = phys.merge(participants, on='event_id')
```

```python
phys_participants
```

```python
phys_participants_quality = phys_participants.merge(quality, left_on='id', right_on='lesson_id', how='inner').drop('id', axis=1)
```

```python
phys_participants_quality
```

```python
res = phys_participants_quality.merge(users, left_on='user_id', right_on='id')
```

```python
res.fillna(0, inplace=True)
```
